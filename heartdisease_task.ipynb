Plotting graphs

```python
import matplotlib.pyplot as plt
x=dataset['thalach']
plt.hist(x)
plt.title('max heart rate achieved-thalach')
```

```python
import matplotlib.pyplot as plt
x=dataset['age']
y=dataset['chol']
plt.scatter(x,y)
plt.title('age-cholestrol')
```

```python
import matplotlib.pyplot as plt
x=dataset['sex']
y=dataset['thal']
plt.bar(x,y)
plt.title('sex-thalassemia')
```

```python
import matplotlib.pyplot as plt
x1=dataset['cp']
y1=dataset['trestbps']
plt.plot(x1,y1,label="chestpain-trestbps")
x2=dataset['cp']
y2=dataset['fbs']
plt.plot(x2,y2,label="chestpain-fbs")

```
